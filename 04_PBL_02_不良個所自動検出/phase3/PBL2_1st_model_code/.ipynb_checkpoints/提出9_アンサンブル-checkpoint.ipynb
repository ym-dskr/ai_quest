```python
import os
import glob
import random
from PIL import Image

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torchvision import transforms as T
from torch.utils.data import DataLoader
import albumentations as A
```

```python
device = torch.device("cuda", 0) if torch.cuda.is_available() else "cpu"
count = torch.cuda.device_count()
def fix_seed(seed):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
SEED = 42
fix_seed(SEED)
```

```python
def make_datapath_list(root_path):
    path_list = []
    target_path = os.path.join("data" + root_path + f'/*.jpeg')
    for path in glob.glob(target_path):
        path_list.append(path)
    path_list.sort()
    path_list = [os.path.basename(p) for p in path_list]
    return path_list
```

```python
model1 = pd.read_csv("csv/eff4.csv")
model2 = pd.read_csv("csv/vit.csv")
model3 = pd.read_csv("csv/res.csv")
```

```python
#15モデルをアンサンブルすればある程度安定したスコアを得ることができます。
#しかし実運用を考えるべきですので、各モデル1つを選んでアンサンブルしています。
model = pd.concat([model1.model1, model3.model3, model3.model1], axis=1)
model
```

```python
f = model.mean(1).values.flatten()
f = np.where(f > 0.5, 1, 0)
```

```python
test = make_datapath_list('/crip_test')
df = pd.DataFrame([test, f]).T
df.to_csv('csv/submission.tsv',
         index=False,
         header=False,
         sep='\t')
```
